```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(open("index.html"), 'lxml')
```

```python
tag = soup.a  # 获取 a 标签
print(tag.name)  # 标签名称：a
print(tag.attrs)  # 属性列表：['href', 'class', 'data']
print(tag.get('href'))  # 获取href属性：index.html
```

```python
comment = soup.div
type(comment)  # <class 'bs4.element.Comment'>
print(comment.p)
```

```python
print(comment.contents)
```

```python
for child in comment.children:
    print(child)
```

```python
for child in comment.descendants:
    print(child)
```
